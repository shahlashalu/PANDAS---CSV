# **TO_CSV**

To convert a file into csv file

```python
#to_csv

import pandas as pd

two = pd.DataFrame({'id':[1,2,3,4,5,7],'Name':['Billy','Brian','Bran','Bryce','Betty','ramu'],'subject_id':['sub2','sub4','sub8','sub6','sub5','sub9'],'Marks_scored':[89,80,79,97,88,56],'class':[1,23,4,5,5,6]},index=[1,2,3,4,5,7])

three = pd.DataFrame({'id':[1,2,3,4,5,6],'Name':['Billy','Brian','Bran','Bryce','Betty','sarang'],'subject_id':['sub2','sub4','sub3','sub6','sub5','sub7'],'Marks_scored':[89,80,79,97,88,76],'roll no':[1,23,4,5,5,6]},index=[1,2,3,4,5,6])
```

```python
df = pd.concat([two,three])
df
```

**TO_CSV**

Covert a file into csv file

```python
df.to_csv('Student.csv')
```
